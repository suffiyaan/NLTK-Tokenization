```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
nltk.download('punkt')
```

```python
text = "Natural Language Processing is a fascinating field of artificial intelligence. It allows machines to understand human language. NLTK is one of the most popular libraries used for text processing in Python. Tokenization is often the first step in NLP pipelines."
```

```python
tokens = word_tokenize(text)
```

```python
print("Tokens:", tokens)
```

```python
num_tokens = len(tokens)
print("Total number of tokens:", num_tokens)
```

```python
frequency_dist = FreqDist(tokens)
print("Token frequencies:", frequency_dist)
```

```python
for token, frequency in frequency_dist.items():
    print(f"'{token}': {frequency}")
```
